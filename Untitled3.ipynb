```python
class Person:
    def __init__(self, ism, familiya, yosh):
        self.ism = ism
        self.familiya = familiya
        self.yosh = yosh
    
    def __str__(self):
        return self.ism
    
    def __repr__(self):
        return f"{self.familiya} - {self.ism}: {self.yosh}"

person = Person("Sanjar","Saidov",20)
print(person)
print(str(person))
print(person.__dict__)
# Sanjar
# Sanjar
# {'ism': 'Sanjar', 'familiya': 'Saidov', 'yosh': 20}
# javoblar shunaqa, savol qo'yvolasmi brat
```
